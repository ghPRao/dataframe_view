# Dataframe View

When you display dataframe in the notebook, you see only partial feature list or partial observations if the features or observations exceeed the display screen.
An option can be set while displaying which will out out all features or observations and the ability to scroll through sliding bar.

```python
import pandas as pd
```

# Read the data

```python
df = pd.read_csv("factorAnalysisData.csv")
```

```python
df
```

# See all columns

```python
pd.set_option('display.max_columns', None) 
# above can be written as pd.set_option('max_columns', None)
df.head()
```

# To reset to defaults

```python
pd.reset_option('max_columns')
df.head()
```

# Set column width to 20

```python
pd.set_option("display.max_colwidth", 20)
```

# Set only 2 columns and all rows

```python
pd.set_option("max_columns", 2) 

pd.set_option("max_rows", None)
df
```

```python
pd.reset_option("display.max_rows", None)
```
